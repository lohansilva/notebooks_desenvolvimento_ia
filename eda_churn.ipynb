## Análise Bivariada

```python
# Hipótese: Clientes com contrato do tipo mensal são mais propensos ao churn
# Construir tabela de contigencia entre tipo de contrato e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir tabela de contigencia entre tipo de contrato e churn - proporção
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, normalize='index',margins_name='Total')
```

considerando o publico total, os clientes de contrato mensal representam 55%. Porem, se considerarmos apenas os clientes que abondonaram o serviço, 88% tinham contrato mensal. Desta forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# avaliar a correlação entre duas variaveis categóricas (qualitativas)

# executar um teste de hipótese chamado chi-square ou qui-quadrado de pearson
# num teste de hipóstese, duas hipóteses são formuladas:
# H0 (Hipótese Nula): as duas variáveis são independentes
# H1 (hipótese complementar): as duas variáveis não são independentes

# O teste serve para confirmar ou recusar a hipótese nula (H0)
# quando a probabilidade de observarmos H0 é inferior a 0.05 (p-value),
# recusamos a hipótese e seguimos com a complementar
```

```python
# gerar um DF da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# calcular os scores e p-values
from scipy.stats import chi2_contingency
```

```python
# tupla = lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# apresentar numeros com decimais sem a notação cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# criar dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# analisar scores e p-values
df_chi_scores_churn_contract
```

Conforme Teste Estatistico (Chi-square), o p-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variaveis não são independentes. Pelo qui2 alto, podemos afirmar que há uma forte correlação.

```python
# avaliar a correlação entre uma variavel qualitativa e quantitativa
# usar o mesmo teste de chi-square, mas transformar a variavel quantitativa em qualitativa
# hipotese: cliente com menos de 6 meses de contrato é mais propenso ao churn

# criar uma coluna nova no dataframe com base numa condição de outra coluna
# coluna 'TempoMenor6Meses'

import numpy as np
```

```python
# criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
df_churn
```

```python
# gerar um DF da Crosstab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# calcular os scores
# tupla = lista imutável
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
```

```python
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

Como o p-value <= 0.05, rejeitamos a H0, ou seja, as variaveis não são independentes. E vale mencionar que com base do qui2 resultante das 2 analises, a correlação entre churn e tempo de contrato < 6 meses é menor do que correlação entre churn e tipo de contrato.

```python
# correlação entre duas variáveis numericas
# tenure com TotalCharges
# A intuição é que quanto mais tempo de contrato maior o valor pago.

#correlação entre 2 variaveis numéricas - pearson
df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
#correlação entre 2 variaveis numéricas - Sperman
df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```

há uma forte correlação entre Tenure e TotalCharges, usando métodos estatisticos de correlação

```python
df_churn_month
```

```python
# histograma - checar visualmente se os dados seguem uma distribuição normal
df_churn_month.TotalCharges.plot.hist()
```

```python
# Metodo de Tukey - IQR (Distribuição não normal)
# IQR - Range Interquartil
# IQR = 3o quartil -1o quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites Inferior e Superior
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

```python
# zscore (usado para distribuição normal)
# indica quantos desvios padrões um ponto específico de dados está distante da média
# zscore = (x - media) / desvio padrão
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
```
