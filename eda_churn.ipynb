```python
# Renomear colunas
df_services.rename(columns={'customerID' : 'IDCliente'}, inplace=True)
```

```python
# Unificar DataFrame de Customers com Services, criando um terceiro DataFrame
df_temp = df_customers.merge(df_services, on=['IDCliente'])
```

```python
df_temp.info()
```

```python
df_temp.head(2)
```

```python
# unificar df_temp com Contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Unificar os três dataframes ao mesmo tempo, com colunas com nomes diferentes
df_churn = df_customers.merge(df_services, on=['IDCliente']).merge(df_contracts, left_on=['IDCliente'], right_on=['customerID'])
```

```python
df_churn.info()
```

```python
# Removendo coluna de um Dataframe
df_churn.drop(['customerID'], axis=1, inplace=True)
```

```python
df_churn.info()
```

```python
# quantos clientes possuem 1 mes de contrato - filtro
len(df_churn[(df_churn.tenure==1)])
```

```python
# quanto os clientes de 1 mes de contrato representam percentualmente
len(df_churn[(df_churn.tenure==1)]) / len(df_churn) * 100
```

```python
# quantos clientes possuem entre 1 e 6 meses de contrato
len(df_churn[(df_churn.tenure>=1) & (df_churn.tenure<=6)])
```

```python
# quantos clientes possuem entre 1 e 6 meses de contrato e são do sexo masculino
len(df_churn[(df_churn.Genero=='Male') & (df_churn.tenure<=6)])
```

```python
# apresentar a quantidade de clientes por tempo de contrato - agrupamento de dados com sumarização
df_churn.groupby(['tenure'])['tenure'].count().sort_values(ascending=False)
```

```python
# apresentar a quantidade de clientes por tempo de contrato - agrupamento de dados com sumarização - em um plot
plot_x = df_churn.groupby(['tenure'])['tenure'].count().sort_values().plot.barh(figsize=(20,20))

plot_x.bar_label(plot_x.containers[0]);
```
