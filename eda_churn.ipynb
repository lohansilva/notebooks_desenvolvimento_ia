## An√°lise Bivariada

```python
# Hip√≥tese: Clientes com contrato do tipo mensal s√£o mais propensos ao churn
# Construir tabela de contigencia entre tipo de contrato e churn - quantidade
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Construir tabela de contigencia entre tipo de contrato e churn - propor√ß√£o
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, normalize='index',margins_name='Total')
```

considerando o publico total, os clientes de contrato mensal representam 55%. Porem, se considerarmos apenas os clientes que abondonaram o servi√ßo, 88% tinham contrato mensal. Desta forma, podemos afirmar que h√° uma correla√ß√£o entre o tipo de contrato mensal e o abandono.

```python
# avaliar a correla√ß√£o entre duas variaveis categ√≥ricas (qualitativas)

# executar um teste de hip√≥tese chamado chi-square ou qui-quadrado de pearson
# num teste de hip√≥stese, duas hip√≥teses s√£o formuladas:
# H0 (Hip√≥tese Nula): as duas vari√°veis s√£o independentes
# H1 (hip√≥tese complementar): as duas vari√°veis n√£o s√£o independentes

# O teste serve para confirmar ou recusar a hip√≥tese nula (H0)
# quando a probabilidade de observarmos H0 √© inferior a 0.05 (p-value),
# recusamos a hip√≥tese e seguimos com a complementar
```

```python
# gerar um DF da Crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# calcular os scores e p-values
from scipy.stats import chi2_contingency
```

```python
# tupla = lista imut√°vel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# apresentar numeros com decimais sem a nota√ß√£o cientifica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# criar dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# analisar scores e p-values
df_chi_scores_churn_contract
```

Conforme Teste Estatistico (Chi-square), o p-value <= 0.05, desta forma rejeitamos a hip√≥tese nula, ou seja, as variaveis n√£o s√£o independentes. Pelo qui2 alto, podemos afirmar que h√° uma forte correla√ß√£o.

```python
# avaliar a correla√ß√£o entre uma variavel qualitativa e quantitativa
# usar o mesmo teste de chi-square, mas transformar a variavel quantitativa em qualitativa
# hipotese: cliente com menos de 6 meses de contrato √© mais propenso ao churn

# criar uma coluna nova no dataframe com base numa condi√ß√£o de outra coluna
# coluna 'TempoMenor6Meses'

import numpy as np
```

```python
# criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
df_churn
```

```python
# gerar um DF da Crosstab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# calcular os scores
# tupla = lista imut√°vel
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
```

```python
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

Como o p-value <= 0.05, rejeitamos a H0, ou seja, as variaveis n√£o s√£o independentes. E vale mencionar que com base do qui2 resultante das 2 analises, a correla√ß√£o entre churn e tempo de contrato < 6 meses √© menor do que correla√ß√£o entre churn e tipo de contrato.

```python
# correla√ß√£o entre duas vari√°veis numericas
# tenure com TotalCharges
# A intui√ß√£o √© que quanto mais tempo de contrato maior o valor pago.

#correla√ß√£o entre 2 variaveis num√©ricas - pearson
df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
#correla√ß√£o entre 2 variaveis num√©ricas - Sperman
df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```

h√° uma forte correla√ß√£o entre Tenure e TotalCharges, usando m√©todos estatisticos de correla√ß√£o

```python
# Metodo de Tukey - IQR (Distribui√ß√£o n√£o normal)
# IQR - Range Interquartil
# IQR = 3o quartil -1o quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
# Limites Inferior e Superior
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) | (df_churn_month.TotalCharges > limSup_TotalCharges_month)]
```

```python
# zscore (usado para distribui√ß√£o normal)
# indica quantos desvios padr√µes um ponto espec√≠fico de dados est√° distante da m√©dia
# zscore = (x - media) / desvio padr√£o
from scipy.stats import zscore
```
